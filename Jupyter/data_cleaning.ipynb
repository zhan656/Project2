Since the file contains multiple language characters, declare the encoding method with read_csv()

```python
# Read CSV file with encoding method 'utf-8' and tranfer it into dataframe
raw = pd.read_csv("../Data/winemag-data_first150k.csv",encoding='utf-8')
raw_df = pd.DataFrame(raw)
happy = pd.read_csv("../Data/happiness_report.csv")
happy_df = pd.DataFrame(happy)
income_raw = pd.read_csv('../Data/average_income.csv')
income_df = pd.DataFrame(income_raw)
pop_raw = pd.read_csv('../Data/world_population.csv')
pop_df = pd.DataFrame(pop_raw)
```

With the dateframe has 150930 rows and 11 columns, we want to remove column "region_2" which contains a large amount of "NaN" data. Also rows that contains NaN values in the remaining dataframe.

```python
semi_df = raw_df.drop(['region_2'],axis=1).dropna()
semi_df.tail(10)
```

```python
semi_df['ID'] = np.arange(len(semi_df["ID"])) # Reset ID Column 
semi_df.tail(10)
```

```python
uniq_country = list(set(semi_df["country"]))
uniq_country # all unique country names, to filter record in the happiness_df
```

```python
happy_ref = {}
for ctry in uniq_country:
    if ctry != 'US':
        happy_ref[ctry] = happy_df.loc[happy_df['Country']==ctry]['Happiness.Score'].tolist()[0]
    else:
        happy_ref[ctry] = happy_df.loc[happy_df['Country']=='United States']['Happiness.Score'].tolist()[0]
```

```python
happy_ref
```

```python
income_ref = {}
```

```python
for ctry in uniq_country:
    if ctry != 'US':
        income_ref[ctry] = income_df.loc[income_df['Country']==ctry]['Average income'].tolist()[0]
    else:
        income_ref[ctry] = income_df.loc[income_df['Country']=='United States']['Average income'].tolist()[0]
income_ref
```

```python
pop_ref = {}
for ctry in uniq_country:
    if ctry != 'US':
        pop_ref[ctry] = pop_df.loc[pop_df['Geography']==ctry]['Pop_mid_2017 (Million)'].tolist()[0]
    else:
        pop_ref[ctry] = pop_df.loc[pop_df['Geography']=='United States']['Pop_mid_2017 (Million)'].tolist()[0]
pop_ref
```

```python
# add a new column call "happy_score" and assign score value from "happy_ref"
semi_df["happy_score"]=''
semi_df['avg_income']=''
for index, row in semi_df.iterrows():
    semi_df.set_value(index, "happy_score", happy_ref[row['country']])
    semi_df.set_value(index, "avg_income", income_ref[row['country']])
semi_df.tail()
```

```python
avg_ctry_df = semi_df.loc[:,['country','points','price']]
avg_ctry_df.head(10)
```

```python
country_df = pd.DataFrame(avg_ctry_df.groupby('country')['points','price'].mean()).reset_index()
country_df
```

```python
winery_ct_ref = {}
variety_ct_ref = {}
for ctry in uniq_country:
    winery_ct_ref[ctry] = len(list(set(semi_df.loc[semi_df['country']==ctry]['winery'])))
    variety_ct_ref[ctry] = len(list(set(semi_df.loc[semi_df['country']==ctry]['variety'])))
winery_ct_ref
```

```python
variety_ct_ref
```

```python
country_df['happy_score'] = ''
country_df['avg_income'] = ''
country_df['winery_count'] = ''
country_df['variety_count'] = ''
country_df['pop_2017_million'] = ''
country_df
```

```python
for index, row in country_df.iterrows():
    country_df.set_value(index, "happy_score", happy_ref[row['country']])
    country_df.set_value(index, "avg_income", income_ref[row['country']])
    country_df.set_value(index, "winery_count", winery_ct_ref[row['country']])
    country_df.set_value(index, "variety_count", variety_ct_ref[row['country']])
    country_df.set_value(index, "pop_2017_million", pop_ref[row['country']])
country_df.tail(10)
```

```python
country_df.to_csv('../Data/country_avg.csv',encoding = 'utf=8',index = False)
```
